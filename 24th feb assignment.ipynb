```python

def reindex_dataframe(df):
    # reset the index to start from 0
    df.reset_index(inplace=True)

    # set a new index that starts from 1 and increments by 2 for each row
    df.index = df.index * 2 + 1

    # return the updated DataFrame
    return df
reindex_dataframe(df)
```

```python
def stats_cal(df):
    mean_value = df['values'].mean()
    median_value = df['values'].median()
    std = df['values'].std()
    
    return{'Mean':mean_value,'Median':median_value,'Std':std}
```

```python
def new_col(df):
    df['Moving_avg'] = df['Sales'].rolling(window = 7, min_periods = 1).mean()
    return df
    
```

```python
#Question 11:

def add_weekday(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.strftime('%A')
    return df
```

```python
df = pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']})
```

```python
df
```

```python
df = add_weekday(df)
```
